```python
"""
Using Python (Pandas), can you sort the dataframe in descending order 
with the most calorically dense food (normalized for weight) at the top?
"""
import pandas as pd
raw_data = {'food': ['bacon', 'strawberries', 'banana', 'spinach', 'chicken breast', 'peanuts'], 
            'grams': [50, 200, 100, 200, 50, 100],
           'calories': [271, 64, 89, 46, 80, 567]}
df = pd.DataFrame(raw_data, columns = ['food', 'grams', 'calories'])
df
```

```python
df['density'] = df.calories / df.grams
df
```

```python
df['density_norm'] = df.density / (df.density.max() - df.density.min())
df
```

```python
idx = df.density_norm.sort_values(ascending = False).index
```

```python
df.iloc[idx]
```
