```python
import pandas as pd
```

```python
df = pd.read_csv('./data/stock_prices.csv')
df.head()
```

```python
df.shape
```

```python
df['7_MVA'] = df['Close'].rolling(7).mean()
df['7_STDev'] = df['Close'].rolling(7).std()
df['high'] = df['Close'] + 1.5 * df['7_STDev']
df['low'] = df['Close'] - 1.5 * df['7_STDev']
df.head(15)
```

```python
df.plot('Date', ['Close', '7_MVA', 'high', 'low'], rot=45)
```

```python
import matplotlib.pyplot as plt
plt.plot(df['7_MVA'])
plt.fill_between(df['Date'], df['low'], df['high'], alpha = 0.1)
;
```
