```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/ddi_message_app_data.csv')
```

```python
data
```

```python
"""
Using Python, calculate what fraction of senders sent messages to at least 9 unique people on March 1, 2018.

"""
data.info()
```

```python
data_date = data[data['date'] == '2018-03-01']
```

```python
data_date.shape
```

```python
data_date['date'].nunique(), data_date['sender_id'].nunique(), data_date['receiver_id'].nunique()
```

```python
tuples = []
for i in range(0, data_date.shape[0]):
    tuples.append((data_date['sender_id'].iloc[i], data_date['receiver_id'].iloc[i]))
len(tuples)
```

```python
tuples
```

```python
tuples_dict = {}
for t in tuples:
    if t in tuples_dict:
        tuples_dict[t] += 1
    else:
        tuples_dict[t] = 1
tuples_dict
```

```python
sender_dict = {}
for x in tuples_dict:
    if x[0] in sender_dict:
        sender_dict[x[0]] += 1
    else:
        sender_dict[x[0]] = 1
sender_dict
```

```python
len(tuples_dict)
```

```python
popular_senders = []
for key in sender_dict:
    if sender_dict[key] >= 9:
        popular_senders.append(key)
```

```python
popular_senders
```

```python
unique_senders = data['sender_id'].nunique()
unique_senders
```

```python
percent_popular = len(popular_senders) / unique_senders
percent_popular
```
