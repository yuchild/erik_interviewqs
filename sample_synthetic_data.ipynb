```python
from sdv import load_demo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
metadata, tables = load_demo(metadata=True)
```

```python
tables
```

```python
users_df = tables['users']
users_df
```

```python
sessions_df = tables['sessions']
sessions_df
```

```python
trans_df = tables['transactions']
trans_df
```

```python
metadata
```

```python
user_sessions = users_df.merge(sessions_df
                               , left_on = 'user_id'
                               , right_on = 'user_id'
                               , how = 'outer'
                              )
user_sessions
```

```python
user_sessions_trans = user_sessions.merge(trans_df
                                          , left_on = 'session_id'
                                          , right_on = 'session_id'
                                          , how = 'outer'
                                         )
user_sessions_trans
```

```python
user_sessions_trans.fillna('None', inplace = True)
user_sessions_trans
```

```python
user_sessions_trans.groupby('country')['age'].mean()
```

```python
user_sessions_trans.groupby(['device', 'approved'])['approved'].count()
```
